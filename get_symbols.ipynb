```python
list_results = []

for comp in tqdm(profit_companies) :
    url_filled = url.format(name=comp)
    response = requests.get(url_filled)
    if (response.status_code == 200) | (response.status_code == 404) :
        response_json = response.json()
        response_json['name'] = comp
        list_results.append(response_json)
        sleep(3)
    else :
        print('request failed :')
        print(response.json())
```

```python
df_results = pd.DataFrame(list_results)
df_results
```

```python
df_count = pd.DataFrame(df_results['stock_symbol'].value_counts() > 1).reset_index()
df_unique = df_count.loc[df_count['count'] == False, ['stock_symbol']]
df_ticker = df_unique.merge(df_results, on=['stock_symbol']).sort_values('stock_symbol').drop(columns=['detail'])
df_ticker
```

```python
df_nifty500 = pd.read_csv('./data/ind_nifty500list.csv')
df_final = df_nifty500.merge(df_ticker, left_on='Symbol', right_on='stock_symbol').sort_values('stock_symbol')
df_final
```
