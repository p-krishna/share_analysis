```python
# Fetch historical data
from nsepython import equity_history

stock_symbol = "RELIANCE"
start_date = "01-01-2019"  # Format: DD-MM-YYYY
end_date = "10-02-2025"    # Format: DD-MM-YYYY

# Fetch data
stock_data = equity_history(
    symbol=stock_symbol,
    start_date=start_date,  # Use 'from_date' instead of 'start'
    end_date=end_date,       # Use 'to_date' instead of 'end'
    series="EQ"             # Equity series
)

# Check if data is empty
if stock_data.empty:
    print("Data is empty. Please check the stock symbol and date range.")
else:
    print("Data fetched successfully:")
    print(stock_data.head())
```
