```python
import os
import pandas as pd
```

```python
data_dir = 'data/usd_to_inr'
```

```python
files = [f for f in os.listdir(data_dir) if (f.startswith('USDINR')) & (f.endswith('.csv'))]
files.sort()
```

```python
df_list = []
for f in files:
    df = pd.read_csv(os.path.join(data_dir, f))
    df_list.append(df)

df = pd.concat(df_list, ignore_index=True)
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df = df.sort_values('Date').reset_index(drop=True)
df_final = df[['Date', 'Close']].drop_duplicates().rename(columns={'Date': 'date', 'Close': 'usd_to_inr'})
```

```python
df_final.to_csv('data/usd_to_inr.csv', index=False)
```
